```python
import requests
import pandas as pd

# Fetch products from Fake Store API
url = "https://fakestoreapi.com/products"
response = requests.get(url)

products = response.json()
products_df = pd.DataFrame(products)

# Show few products
products_df[['id', 'title', 'price', 'category']].head()
```

```python
# Create a text field to embed
products_df['text'] = (
    products_df['title'] + " " +
    products_df['description'] + " Category: " +
    products_df['category']
)

# Quick check
products_df[['text']].head(2)
```

```python
from sentence_transformers import SentenceTransformer

# Load a lightweight embedding model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Embed the product texts
embeddings = embedder.encode(products_df['text'].tolist(), show_progress_bar=True)
```

```python
import faiss
import numpy as np

# Convert to float32
embeddings = np.array(embeddings).astype('float32')

# Build FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

print(f"FAISS index size: {index.ntotal}")
```

```python
def search_products(query, top_k=5):
    # Embed the query
    query_vector = embedder.encode([query]).astype('float32')

    # Search FAISS
    distances, indices = index.search(query_vector, top_k)

    # Fetch matching products
    results = products_df.iloc[indices[0]]
    return results[['id', 'title', 'price', 'category', 'description']]

# Example search
search_products("Nike sneakers under $120")
```

```python
import re

def parse_query(user_query):
    parsed = {}

    # Detect price
    price_match = re.search(r'under \$?(\d+)', user_query)
    if price_match:
        parsed['max_price'] = int(price_match.group(1))

    # Detect simple categories
    categories = ['shoes', 't-shirt', 'shirt', 'sneakers', 'bag', 'jewelry', 'electronics']
    for cat in categories:
        if cat in user_query.lower():
            parsed['category'] = cat

    # Always store original query
    parsed['query'] = user_query

    return parsed

# Example
parse_query("find sneakers under $120")
```

```python
def smart_search(user_query, top_k=5):
    parsed = parse_query(user_query)

    # Embed and search
    query_vector = embedder.encode([parsed['query']]).astype('float32')
    distances, indices = index.search(query_vector, top_k*2)  # fetch extra in case filters cut

    # Fetch results
    results = products_df.iloc[indices[0]]

    # Apply price filter
    if 'max_price' in parsed:
        results = results[results['price'] <= parsed['max_price']]

    # Apply category filter
    if 'category' in parsed:
        results = results[results['category'].str.contains(parsed['category'], case=False, na=False)]

    return results[['id', 'title', 'price', 'category', 'description']].head(top_k)

# Try
smart_search("Find sneakers under $120")
```

```python
def suggest_bundle(base_product_title, top_k=3):
    # Embed the base product title
    query_vector = embedder.encode([base_product_title + " accessories"]).astype('float32')

    # Search
    distances, indices = index.search(query_vector, top_k)

    # Fetch results
    results = products_df.iloc[indices[0]]
    return results[['id', 'title', 'price', 'category']]

# Example
suggest_bundle("sneakers")
```

```python
from langchain.agents import tool

@tool
def search_products_tool(query: str) -> str:
    """Search products based on query (supports price and category)."""
    results = smart_search(query)
    return results.to_string(index=False)

@tool
def suggest_bundle_tool(product_title: str) -> str:
    """Suggest accessories or matching products for a given product."""
    results = suggest_bundle(product_title)
    return results.to_string(index=False)
```

```python
from langchain.agents import Tool, initialize_agent, AgentType
from langchain.prompts import PromptTemplate
from langchain_huggingface import HuggingFaceEndpoint
from langchain.chains import LLMChain
import os

# Initialize HuggingFace model (gpt2) for text-generation task
llm = HuggingFaceEndpoint(
    repo_id="gpt2",  # The model we are using
    task="text-generation",  # Specify the task as 'text-generation'
    temperature=0.5,
    max_new_tokens=512,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

# Define your tools using LangChain's Tool class
def search_products(query):
    # Sample mock-up, in a real scenario, this could integrate with a product database or API
    if "sneakers" in query.lower():
        return "Found sneakers under $120: Nike Air Zoom, Adidas Ultraboost. Would you like to check them?"
    return "Sorry, I couldn't find any sneakers under $120."

def suggest_socks(shoes_name):
    # Based on selected shoes, suggest socks
    if "Nike Air Zoom" in shoes_name:
        return "Matching socks for Nike Air Zoom: Nike Elite Crew Socks, Adidas Performance Cushioned Socks."
    elif "Adidas Ultraboost" in shoes_name:
        return "Matching socks for Adidas Ultraboost: Under Armour HeatGear Socks, Puma Sports Cushioned Socks."
    return "Sorry, no sock recommendations available."

# Wrap functions into tools
search_products_tool = Tool(
    name="Search Products",
    func=search_products,
    description="Search for products based on a query"
)

suggest_socks_tool = Tool(
    name="Suggest Socks",
    func=suggest_socks,
    description="Suggest matching socks based on a pair of shoes"
)

# Now use these tools for the agent
tools = [search_products_tool, suggest_socks_tool]

# Define the prompt template for text generation task
prompt_template = "User input: {input}\nResponse:"
prompt = PromptTemplate(input_variables=["input"], template=prompt_template)

# Chain the model with the prompt using LLMChain
llm_chain = LLMChain(prompt=prompt, llm=llm)

# Initialize the agent with the correct parameters
agent = initialize_agent(
    tools=tools,  # Use the tools you've defined
    llm=llm_chain,  # Properly linking LLM with the tools
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,  # Use a zero-shot agent type
    verbose=True  # Enable verbose to see detailed outputs
)

# Full user interaction workflow:
query = "Find me sneakers under $120 and suggest matching socks."

# Search products and suggest socks
search_response = search_products(query)  # First get sneakers
print("Sneaker Search:", search_response)

# If sneakers are found, suggest socks
if "Nike Air Zoom" in search_response:
    socks_response = suggest_socks("Nike Air Zoom")
    print("Sock Suggestions:", socks_response)
elif "Adidas Ultraboost" in search_response:
    socks_response = suggest_socks("Adidas Ultraboost")
    print("Sock Suggestions:", socks_response)
else:
    print("No matching sneakers found.")
```

```python
# Step 1: Display a list of sneakers under $120
sneakers = {
    "Nike Air Zoom": 110,
    "Adidas Ultraboost": 100,
    "Puma RS-X": 90,
    "Reebok Nano": 95,
}

# Function to display sneakers
def display_sneakers(sneakers):
    print("Sneakers under $120:")
    for idx, (sneaker, price) in enumerate(sneakers.items(), 1):
        print(f"{idx}. {sneaker} - ${price}")
    print("Select a sneaker by number (1-4):")

# Get the user's selection
def select_sneaker(sneakers):
    display_sneakers(sneakers)
    selection = int(input("Enter the sneaker number: "))
    selected_sneaker = list(sneakers.keys())[selection - 1]
    return selected_sneaker

# Call the function to get the selected sneaker
selected_sneaker = select_sneaker(sneakers)
print(f"You selected: {selected_sneaker}")
```

```python
# Step 2: Matching socks for the selected sneaker
def suggest_socks(sneaker):
    sock_suggestions = {
        "Nike Air Zoom": ["Nike Elite Crew Socks", "Nike Sports Performance Socks"],
        "Adidas Ultraboost": ["Adidas Performance Cushioned Socks", "Adidas No-Show Socks"],
        "Puma RS-X": ["Puma No-Show Socks", "Puma Performance Crew Socks"],
        "Reebok Nano": ["Reebok Performance Socks", "Reebok Cushioned Crew Socks"]
    }

    # Return sock suggestions based on the selected sneaker
    return sock_suggestions.get(sneaker, [])

# Get sock suggestions
socks = suggest_socks(selected_sneaker)
print("\nSuggested socks:")
for sock in socks:
    print(f"- {sock}")
```

```python
# Complete flow: Display sneakers, allow selection, and suggest socks
def run_interactive_flow():
    sneakers = {
        "Nike Air Zoom": 110,
        "Adidas Ultraboost": 100,
        "Puma RS-X": 90,
        "Reebok Nano": 95,
    }

    selected_sneaker = select_sneaker(sneakers)
    print(f"\nYou selected: {selected_sneaker}")

    socks = suggest_socks(selected_sneaker)
    print("\nSuggested socks:")
    for sock in socks:
        print(f"- {sock}")

# Run the flow
run_interactive_flow()
```

```python
class SneakerShoppingAgent:
    def __init__(self):
        # Define sneakers and sock options with prices
        self.sneakers = [
            {"name": "Nike Air Zoom", "price": 110},
            {"name": "Adidas Ultraboost", "price": 100},
            {"name": "Puma RS-X", "price": 90},
            {"name": "Reebok Nano", "price": 95}
        ]
        self.socks = {
            "Nike Air Zoom": [{"name": "Nike Elite Crew Socks", "price": 20}, {"name": "Adidas Performance Cushioned Socks", "price": 18}],
            "Adidas Ultraboost": [{"name": "Adidas Cushioned Performance Socks", "price": 22}, {"name": "Nike Elite Crew Socks", "price": 20}],
            "Puma RS-X": [{"name": "Puma Performance Crew Socks", "price": 15}, {"name": "Reebok Cushioned Crew Socks", "price": 18}],
            "Reebok Nano": [{"name": "Reebok Performance Socks", "price": 17}, {"name": "Reebok Cushioned Crew Socks", "price": 16}]
        }

    def show_sneakers(self):
        # Display available sneakers with prices
        print("Sneakers under $120:")
        for i, sneaker in enumerate(self.sneakers):
            print(f"{i + 1}. {sneaker['name']} - ${sneaker['price']}")

    def show_socks(self, selected_sneaker):
        # Display suggested socks for selected sneaker
        print("\nSuggested socks:")
        for sock in self.socks[selected_sneaker]:
            print(f"- {sock['name']} - ${sock['price']}")

    def get_user_choice(self, prompt):
        # Handle user choice with error checking
        try:
            choice = int(input(prompt))
            return choice
        except ValueError:
            print("Please enter a valid number.")
            return self.get_user_choice(prompt)

    def shopping_cart(self, sneaker_choice, sock_choice):
        # Simulate adding to a shopping cart
        sneaker = self.sneakers[sneaker_choice - 1]
        selected_sneaker = sneaker["name"]
        selected_socks = self.socks[selected_sneaker][sock_choice - 1]

        print("\nYour cart:")
        print(f"{selected_sneaker} - ${sneaker['price']}")
        print(f"{selected_socks['name']} - ${selected_socks['price']}")
        total_price = sneaker["price"] + selected_socks["price"]
        print(f"Total: ${total_price}")

        proceed = input("\nWould you like to proceed with the purchase? (yes/no): ").strip().lower()
        if proceed == 'yes':
            print("Thank you for your purchase!")
        else:
            print("You can continue browsing.")

    def run(self):
        # Show sneakers and get user choice
        self.show_sneakers()
        sneaker_choice = self.get_user_choice("\nSelect a sneaker by number (1-4): ")

        selected_sneaker = self.sneakers[sneaker_choice - 1]["name"]
        self.show_socks(selected_sneaker)

        sock_choice = self.get_user_choice(f"\nSelect a sock by number (1-{len(self.socks[selected_sneaker])}): ")

        # Display cart and ask for confirmation
        self.shopping_cart(sneaker_choice, sock_choice)

# Run the shopping agent
shopping_agent = SneakerShoppingAgent()
shopping_agent.run()
```

```python
import requests

# Function to fetch products from the Fake Store API
def get_products():
    response = requests.get("https://fakestoreapi.com/products")
    return response.json()  # Parse the JSON data into a Python list

# Function to display the products
def display_products(products):
    print("\nAvailable Products:")
    for i, product in enumerate(products, 1):
        print(f"{i}. {product['title']} - ${product['price']} (Category: {product['category']})")

# Function to display the cart content
def display_cart(cart):
    total_amount = 0
    print("\nYour Cart:")
    for item in cart:
        print(f"{item['name']} - ${item['price']}")
        total_amount += item['price']
    print(f"Total: ${total_amount}")
    return total_amount

# Function to simulate checkout process
def checkout(user, cart):
    total_amount = display_cart(cart)
    print(f"\nProceeding with the checkout...\n")
    print(f"User: {user['name']} | Email: {user['email']}")
    print(f"Total amount: ${total_amount}")
    confirmation_number = "123456"  # Mock confirmation number
    print(f"Payment Successful! Your confirmation number is: {confirmation_number}")

# User login function
def user_login():
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    user = {"name": name, "email": email}
    print(f"\nProfile created for {name}!")
    return user

# Main function to control the shopping flow
def main():
    user = user_login()

    # Fetch products from Fake Store API
    products = get_products()

    # Allow filtering by category (optional)
    filter_choice = input("Would you like to filter products by category? (yes/no): ").strip().lower()
    if filter_choice == "yes":
        category = input("Enter the category to filter by (e.g., electronics, jewelery, etc.): ").strip().lower()
        filtered_products = [product for product in products if category in product['category'].lower()]
        if filtered_products:
            display_products(filtered_products)
        else:
            print("No products found in this category.")
            display_products(products)
    else:
        display_products(products)

    # Shopping cart
    cart = []
    while True:
        try:
            product_choice = int(input("Select a product by number: "))
            if 1 <= product_choice <= len(filtered_products if filter_choice == "yes" else products):
                selected_product = (filtered_products if filter_choice == "yes" else products)[product_choice - 1]
                cart.append({"name": selected_product['title'], "price": selected_product['price']})
            else:
                print("Invalid selection, please choose a valid product number.")
                continue
        except ValueError:
            print("Invalid input, please enter a number.")
            continue

        # Display the cart
        total_amount = display_cart(cart)

        # Ask if the user wants to proceed to checkout or continue shopping
        checkout_choice = input("Would you like to proceed with checkout? (yes/no): ").strip().lower()
        if checkout_choice == "yes":
            checkout(user, cart)
            break
        else:
            print("Continuing shopping...")

# Run the main function to start the shopping process
if __name__ == "__main__":
    main()
```

```python
import random
import time
import requests

# Fetch products from the Fake Store API
def fetch_products():
    url = "https://fakestoreapi.com/products"
    response = requests.get(url)
    return response.json()  # Return the product data as a JSON

# Function to display available products
def show_products(products):
    print("Available Products:")
    for i, product in enumerate(products, 1):
        print(f"{i}. {product['title']} - ${product['price']} (Category: {product['category']})")

# User profile creation
def create_profile():
    print("Welcome to the store!")
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    return {"name": name, "email": email}

# Cart management
def manage_cart(products):
    cart = []
    while True:
        show_products(products)
        choice = int(input("Select a product by number (or 0 to finish): "))
        if choice == 0:
            break
        if 1 <= choice <= len(products):
            cart.append(products[choice-1])
            print(f"Added {products[choice-1]['title']} to your cart.")
        else:
            print("Invalid choice. Try again.")

    return cart

# Checkout process
def checkout(cart, user_info):
    if not cart:
        print("Your cart is empty!")
        return

    print("\nYour Cart:")
    total = 0
    for item in cart:
        print(f"{item['title']} - ${item['price']}")
        total += item['price']

    print(f"Total: ${total}")

    proceed = input("\nWould you like to proceed with checkout? (yes/no): ")
    if proceed.lower() == "yes":
        confirmation_number = generate_confirmation_number()
        print(f"\nProceeding with checkout...\n")
        print(f"User: {user_info['name']} | Email: {user_info['email']}")
        print(f"Total amount: ${total}")
        print(f"Payment Successful! Your confirmation number is: {confirmation_number}")
    else:
        print("Checkout canceled.")

# Generate a unique confirmation number
def generate_confirmation_number():
    return random.randint(10000000000000, 99999999999999)

# Main shopping process
def start_shopping():
    products = fetch_products()  # Fetch products from Fake Store API
    user_info = create_profile()  # Create user profile
    cart = manage_cart(products)  # Manage cart
    checkout(cart, user_info)  # Proceed with checkout

# Start the shopping process
start_shopping()
```
